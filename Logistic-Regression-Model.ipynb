```python
import pandas as pd
```

```python
#read the csv file and save it in a variable called fraudbank_df
fraudbank_df = pd.read_csv('fraudTest-CSV.csv', sep=',')
```

```python
fraudbank_df
```

```python
#Make a dictionary that assigns a number to each category
#Map that dictionary into a new column of name: "category_num"

diccionario = {'travel': 1, 'shopping_pos': 2, 'shopping_net': 3, 'personal_care': 4, 'misc_pos': 5, 'misc_net': 6, 'kids_pets': 7, 'home': 8, 'health_fitness': 9, 'grocery_pos': 10, 'grocery_net': 11, 'gas_transport': 12, 'food_dining': 13, 'entertainment': 14}
fraudbank_df['category_num']= fraudbank_df['category'].map(diccionario)

#We do a .head to see if everything is fine
fraudbank_df.head(4)
```

```python
def gender_bin(gender):
    if (gender == "M"):
        return 1
    else:
        return 0
```

```python
fraudbank_df['gender_bin'] = fraudbank_df['gender'].apply(gender_bin)
```

```python
fraudbank_df.head()
```

LOGISTIC REGRESSION MODEL

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
```

```python
x=fraudbank_df[['category_num', 'amt', 'gender_bin','zip','lat', 'long']]

y=fraudbank_df[['is_fraud']]
```

```python
#split de los datos
X_train, X_test, y_train, y_test= train_test_split(x,y,test_size=0.30,random_state=101)
```

```python
logisticmodel = LogisticRegression() #create the model
learn_data = logisticmodel.fit(X_train,y_train)   #learn from data
```

```python
predictions= logisticmodel.predict(X_test)
```

```python
X_test
```

```python
predictions
```

```python
probabilities= learn_data.predict_proba(X_test)
```

```python
probabilities
```

```python
from sklearn.metrics import classification_report 
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score, roc_curve, auc
import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle
from sklearn.metrics import f1_score, precision_score, recall_score, roc_auc_score, roc_curve, auc
from scipy import interp
```

```python
accuracy_score(y_test,predictions)
```

```python
precision_score(y_test,predictions, pos_label=1)
```

```python
recall_score(y_test,predictions, pos_label=1)
```

```python
confusion_matrix(y_test, predictions)
#tru target.  No Fraude   154108    11953
#             Fraude      157       498
#                       No fraude  Fraude

#                    test target
```

```python
 
fpr, tpr, threshold = roc_curve(y_test, probabilities[:,1], pos_label=1)
roc_auc = auc(fpr, tpr)
 
plt.figure()
plt.plot(fpr, tpr, color='orange', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0,1], color='navy',  linestyle='--')
 
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
 
plt.show()
 
display()
```
